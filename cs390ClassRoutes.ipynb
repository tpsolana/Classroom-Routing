```python
#function to calculate distance from room to room 

def dijkstra(graph, start, end):
    # Priority queue to store (distance, node)
    pq = [(0, start)]
    distances = {node: float('inf') for node in graph}
    previous_nodes = {node: None for node in graph}
    distances[start] = 0

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        # If we reached the destination, build and return the path
        if current_node == end:
            path = []
            while current_node:
                path.append(current_node)
                current_node = previous_nodes[current_node]
            return path[::-1], distances[end]  # Reverse the path for correct order

        # Explore neighbors
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # If a shorter path is found
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous_nodes[neighbor] = current_node
                heapq.heappush(pq, (distance, neighbor))

    return None, float('inf')  # No path found

# Example Usage:
start_room = "Entrance"
end_room = "room160"
path, distance = dijkstra(BH1_floorplan, start_room, end_room)

if path:
    print(f"Shortest path from {start_room} to {end_room}: {' -> '.join(path)} (Distance: {distance})")
else:
    print(f"No path found from {start_room} to {end_room}.")
```
